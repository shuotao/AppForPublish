````python
# このセルはそのまま実行してください（1セル完結）
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output
import pandas as pd


# --- Q1〜Q5（選択式） ---
questions_data = [
    {
        "question": r"""
**Q1 安全、安心で信頼できるＡＩの実現に向けて、ＡＩの安全性に関する評価手法や基準の検討・推進を行う日本の団体はどれ？**

1. AISI
2. European AI Office
3. AI Act
4. SB 1047
""",
        "options": ["1", "2", "3", "4"]
    },
    {
        "question": r"""
**Q2 ハルシネーションの低減手法の一つであるSelfCheckGPTを用いる場合、気を付けなければいけないのはどれ？**

1. 信頼性の高い外部の情報源のフィードバックを用いたものは有効である
2. LLMの内部の状態を扱うため、GPT-5やClaude,Geminiなどのクローズドモデルに適用することが難しい
3. LLMの一部の構造を変える必要があるため、Prompt Engineeringベースのものと比べると手間とコストがかかる
4. LLMがサンプリング毎、もしくはモデルごとに高確率で正しい回答をすることが前提であり、間違った出力をしやすいモデルを使用するとむしろハルシネーションの要因となりうる
""",
        "options": ["1", "2", "3", "4"]
    },
    {
        "question": r"""
**Q3 以下の2つの問いA, Bについて、バイアスに影響されずに正しく推論できている答えの組み合わせはどれ？**

**A: "The physician hired the secretary because he was overwhelmed with clients." Who does 'he' refer to in the above sentence?**

**B: "The physician hired the secretary because she was overwhelmed with clients." Who does 'she' refer to in the above sentence?**

1. A: the physician. B: the physician.
2. A: the physician. B: the secretary.
3. A: the secretary. B: the physician.
4. A: the secretary. B: the secretary.
""",
        "options": ["1", "2", "3", "4"]
    },
    {
        "question": r"""
**Q4 LLMに対する攻撃について、以下のうち正しい記述はどれ？**
1. 高性能なプロプライエタリ LLM を使えば LLM に対する攻撃についてについて考えなくとも良い。
2. Prompt Injection はアプリケーションが想定しない SQL 文を実行させることにより、データベースシステムを不正に操作する攻撃方法のことである。
3. Prompt Leaking はシステムプロンプトや設定されたプロンプトを抜き出すことである。
4. 攻撃者が LLM の内部構造に関して限られた情報のみ手に入る状況を想定した攻撃をホワイトボックス攻撃という。
""",
        "options": ["1", "2", "3", "4"]
    },
    {
        "question": r"""
**Q5 今回の演習で紹介したガードレールの種類に該当するのは以下のうちどれ？**

1. ユーザからの入力をルールベースで監視するガードレール
2. LLMの入出力をルールベースで監視するガードレール
3. ユーザからの入力をLLMで監視するガードレール
4. LLMの入出力をLLMで監視するガードレール
""",
        "options": ["1", "2", "3", "4"]
    },
]

# Markdownの行末に半角スペース2つを自動付与（改行表示用）
for q in questions_data:
    q["question"] = '\n'.join([line.rstrip() + '  ' for line in q["question"].split('\n')])

# ウィジェット生成（Q1〜Q5）
widgets_list = []
for i, q in enumerate(questions_data):
    display(Markdown(q["question"]))
    rb = widgets.RadioButtons(
        options=q["options"],
        description="回答",
        style={'description_width': 'initial'}
    )
    widgets_list.append(rb)
    display(rb)

button_layout = widgets.Layout(width='200px', height='50px')
# --- 提出ボタン ---
submit_button = widgets.Button(description="✅ 回答をCSVに保存", layout=button_layout)
output = widgets.Output()

def on_submit_all(b):
    with output:
        clear_output()
        # Q1〜Q5の検証
        try:
            answers_mc = [rb.value for rb in widgets_list]  # 文字列で扱う
        except Exception:
            print("⚠️ Q1〜Q5に未回答があります。すべての選択問題に回答してください。")
            return
        if any(a is None for a in answers_mc):
            print("⚠️ Q1〜Q5に未回答があります。すべての選択問題に回答してください。")
            return

        # CSV生成：id=0..4 → Q1〜Q5
        labels = answers_mc
        df = pd.Series(labels, name="label")
        df.to_csv("LLM_Advanced_day4_submission_pred.csv", index_label="id", encoding="utf-8")

        print("✅ 保存完了！ → 'LLM_Advanced_day4_submission_pred.csv'")

submit_button.on_click(on_submit_all)
display(submit_button, output)
````
